```python
from matplotlib import pyplot as plt
x = [3,4,2,6,2]
y = [19,2,4,2,5]
plt.plot(x, y)
plt.show()
```

```python

plt.bar(x, y)
plt.show()
```

```python

plt.hist(y)
plt.show()
```

```python

x2 = [3,5,6]
y2 = [5,6,6]
plt.scatter(x, y) 
plt.plot(x, y, label='First plot')
plt.plot(x2, y2, label='Second plot')
plt.xlabel('time')
plt.ylabel('price')
plt.title('Demo of matplotlib')
plt.show()
```
