```python
import matplotlib
slices= [1,10]
# show plots inline
%matplotlib inline
# use ggplot style
matplotlib.style.use('ggplot')


eating = [23,4,6]
sleeping = [10,2,33]
activities = ['sleeping', 'eating']
cols = ['g','y']
plt.pie(slices,
        labels=activities,
        colors=cols,
        shadow=True, 
        startangle=90,
        explode=(0.1, 0),
        autopct='%1.1f%%'
       )
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy
import matplotlib
# use ggplot style
matplotlib.style.use('ggplot')

labels = 'Frogs', 'Hogs', 'Dogs'
sizes = numpy.array([5860, 677, 3200])
colors = ['yellowgreen', 'gold', 'lightskyblue']

def absolute_value(val):
    a  = numpy.round(val/100*sizes.sum(), 0)
    return a

plt.pie(sizes,
        labels=labels,
        colors=colors,
        autopct=absolute_value,
        shadow=True)

plt.axis('equal')
plt.show()
```

## CSV file

```python
import csv
x=[]
y=[]
with open('example.txt', 'r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))
plt.plot(x, y, label='Loaded from a file')
plt.show()
```

```python
import numpy as np
x, y = np.loadtxt('example.txt', delimiter=',', unpack=True)
plt.plot(x, y, label='Loaded from a file')
plt.show()
```
